```python
# Graphique 1

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Charger les données à partir du tableau
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

# Calculer la moyenne par profession pour la durée et la qualité du sommeil
variables = ['Sleep Duration', 'Quality of Sleep']
moyennes_par_profession = df.groupby('Occupation')[variables].mean()

# Normaliser les données de la qualité du sommeil sur une échelle de 0 à 8

# Créer la figure p
plt.figure(1, figsize=(16, 6))
bar_width = 0.3

# Position des barres sur l'axe x
indices = np.arange(len(moyennes_par_profession))

# Créer les barres pour chaque variable avec les couleurs spécifiées
colors = ['#9fc5e8', 'orange']
for i, variable in enumerate(variables):
    plt.bar(indices + i * bar_width, moyennes_par_profession[variable], bar_width, label=variable, color=colors[i])

#Nommer les axes
plt.xlabel('Profession', fontweight='bold')
plt.ylabel('Moyenne', fontweight='bold')
plt.title('Moyenne de la durée et de la qualité du sommeil par profession', fontweight='bold')
# Rotation des étiquettes
plt.xticks(indices + bar_width / 2, moyennes_par_profession.index, rotation=90) 

# Tronquer les données de l'axe y pour commencer à 3


# Annoter les barres avec les valeurs moyennes
for i, variable in enumerate(variables):
    for j, moyenne in enumerate(moyennes_par_profession[variable]):
        plt.text(indices[j] + i * bar_width, moyenne + 0.1, str(round(moyenne, 1)), ha='center')

# Légende
plt.legend()

# Enregistrer le graphique comme une photo
plt.savefig('graphique1.png')

# Afficher le graphique
plt.show()

```

```python
# Graphique 2

import pandas as pd
import matplotlib.pyplot as plt

# Charger les données à partir du tableau
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

# Calculer la moyenne du niveau de stress par profession
moyenne_stress = df.groupby('Occupation')['Stress Level'].mean()

# Créer la figure
plt.figure(1, figsize=(12, 6))
bar_width = 0.5

# Créer les barres avec les moyennes du niveau de stress
plt.bar(moyenne_stress.index, moyenne_stress, color="purple", width=bar_width)

# Tronquer les données de l'axe y pour commencer à 3
plt.ylim(bottom=3)

# Nommer les axes
plt.xlabel('Profession', fontweight='bold')
plt.ylabel('Niveau de stress', fontweight='bold')
plt.title('Moyenne du niveau de stress par profession', fontweight='bold')
# Tourner les étiquettes
plt.xticks(rotation=90)

# Annoter les barres avec les valeurs moyennes
for i, moyenne in enumerate(moyenne_stress):
    plt.text(i, moyenne + 0.1, str(round(moyenne, 1)), ha='center', color='black') 

# Enregistrer le graphique comme une photo
plt.savefig('graphique2.png')

# Afficher le graphique
plt.show()
```
